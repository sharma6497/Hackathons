```python
df['WeekEnding_Date']=pd.to_datetime(df['WeekEnding_Date'])
```

```python
df1=df.set_index('WeekEnding_Date')
```

```python
df1=df1.sort_index()
```

```python
df1=df1[['Average']].resample('W').mean()
```

```python
df1['Average']=df1['Average'].fillna(method='ffill')
```

```python
df1['Average'].plot(kind='line')
```

```python
df1['year'] = [d.year for d in df1.index]
df1['month'] = [d.strftime('%b') for d in df1.index]
```

```python
sns.lineplot(y=df1['Average'],x=df1['month'],ci=None,hue=df1['year'])
plt.legend(bbox_to_anchor=(1.02, 0.15), loc='upper left', borderaxespad=0)
```

```python
sns.boxplot(x='year', y='Average', data=df1)#
plt.xticks(rotation=90)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
result_add=seasonal_decompose(x=df1['Average'],model='additive')
result_add.plot()
plt.show()
```

```python
result_mul=seasonal_decompose(x=df1['Average'],model='multiplicative')
result_mul.plot()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller, kpss
```

```python
result_ad=adfuller(df1['Average'])
print('p-value:',result_ad[1])
```

```python
df1['diff']=df1['Average']-df1['Average'].shift(1)
```

```python
df1['Seasonal First Difference']=df1['Average']-df1['Average'].shift(12)
```

```python
df1.isna().sum()
```

```python
result_ad=adfuller(df1['diff'].dropna())
print('p-value:',result_ad[1])
```

```python
result_kpss=kpss(df1['diff'].dropna())
print('p-value:',result_kpss[1])
```

```python
df1['diff'].plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = plot_acf(df1['Seasonal First Difference'].iloc[13:],ax=ax1)
ax2 = fig.add_subplot(212)
fig = plot_pacf(df1['Seasonal First Difference'].iloc[13:],ax=ax2)
```

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model=ARIMA(df1['Average'],order=(0,1,0))
model_fit=model.fit()
```

```python
df1['forecast']=model_fit.predict(start=550,end=800,dynamic=True)
df1[['Average','forecast']].plot()
```

Arima model did not make good predictions,since it is a seasonal data lets try with sarimax.

```python
import statsmodels.api as sm
```

```python
model=sm.tsa.statespace.SARIMAX(df1['Average'],order=(2,1,6),seasonal_order=(2,1,6,12),trend='ct')
results=model.fit()
```

```python
df1['forecast']=results.predict(start=500,end=580,dynamic=True)
df1[['Average','forecast']].plot(figsize=(12,8))
```

```python
test=df1.tail(20)
```

```python
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(test['Average'],test['forecast']))
```
